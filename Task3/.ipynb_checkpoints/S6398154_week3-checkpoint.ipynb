```python
from scipy.constants import G
masses = [1.9891e+30, 1.8986e+27, 5.6846e+26, 
          1.0243e+26, 8.6810e+25, 5.9736e+24, 
          4.8685e+24, 6.4185e+23, 3.3022e+23, 
          7.349e+22, 1.25e22]
radii  = [696340, 71492, 60268, 24764, 
          25559, 6378.1, 6051.8, 3396.2, 
          2439.7, 1738.1, 1195]
names  = ['Sun', 'Jupiter', 'Saturn', 'Neptune', 
          'Uranus', 'Earth', 'Venus', 'Mars', 
          'Mercury', 'Moon', 'Pluto']

vel=[]
for i in range(len(masses)):
    v=(2*G*masses[i]/radii[i])**(1/2)
    vel.append(v)
    print(f"The mass of {names[i]} is {masses[i]:2.2E}")
    
planets={}
for i in range(len(masses)):
    planets[names[i]]=vel[i]

print(planets)
```

```python
sumser=[1]
N=100
for i in range(2,N+1):
    sumser.append(sumser[-1]+(1/(i**2)))

if sumser[-1]-sumser[-2]<sumser[-2]-sumser[-3]:
    print("The series converges")
else:
    print("The series doesn't converge")
```

```python
#3.7
from math import atan,degrees
def angsize(d,D):
    deltarad=2*atan(d/2/D)
    delta=degrees(deltarad)
    return delta

print(angsize(1,1))
```

```python
#3.10
from scipy.constants import pi
r=int(input('r: '))
h=int(input('h: '))
def cylinder(r,h):
    A=2*pi*r**2+2*pi*r*h
    V=pi*r**2*h
    return A,V
area, volume=cylinder(r,h)
ratio=area/volume
print(ratio)
```

```python
#3.17
import numpy as np
def stdev(x):
    means=x.mean()
    a=(x-means)**2
    dev=np.sqrt(a.sum()/len(x))
    return dev
test=np.linspace(0,3,20)
print(stdev(test))
print(np.std(test))
def stdev2(x):
    return dev=np.sqrt(((x-(np.zeros(len(x))+x.mean()))**2).sum()/len(x))
print(stdev2(test))
```

```python
#3.18
import numpy as np
data=np.array([[18.49, 2094.75],
[35.48, 2320.87],
[35.48, 1224.9],
[11.34, 872.13],
[7.12,  432.2],
[9.82,  751],
[23.66, 1410],
[44 ,   2700],
[66.7,  5292],
[52.4, 2550],
[55 ,  5253],
[25.6 , 2419],
[26.19, 2158]])
x=data[:,0]
y=data[:,1]
b=((x-x.mean())*(y-y.mean())).sum()/(((x-x.mean())**2).sum())
a=y.mean()-b*x.mean()
print(r"The fitted equation is $y={:2.2f}+{:2.2f}x$".format(a,b))
```
