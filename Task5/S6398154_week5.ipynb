```python
#5.9
from imageio.v2 import imread
import matplotlib.pyplot as plt
data=imread('m101BW.jpg')

plt.figure(figsize=(10,10))
plt.subplot(3,1,1)
plt.imshow(data, cmap='jet')

plt.subplot(3,1,2)
plt.contour(data,levels=np.percentile(data,[90,95,99]),colors=['b','g','r'])
plt.imshow(data)

plt.subplot(3,1,3)
lim=np.percentile(data,90)
datamask=np.where(data>lim,data,0)

plt.imshow(datamask)
plt.show()
```

```python
#5.12

class Galaxy:    
    """Class for redshift calculation of galaxies"""  
        
    caH_0 = 3968.5      # Calcium H line in rest frame
    from scipy.constants import c
    def __init__(self, name, calciumHline=caH_0):
        self.name = name
        self.caH = calciumHline
    def redshift(self):        
        z = (self.caH - self.caH_0)/self.caH
        return z
    def velocity(self):
        v = self.redshift()*c/1000
        return v

gal=Galaxy('Test',4000)
print(gal.velocity())
```
